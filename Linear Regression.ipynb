```python
from pyspark.ml.regression import LinearRegression
```

```python
data=spark.read.csv("/Volumes/workspace/default/nithish-storage/Ecommerce_Customers.csv",inferSchema=True,header=True)
data.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler
```

```python
assemble=VectorAssembler(inputCols=['Avg Session Length','Time on App','Time on Website','Length of Membership','Yearly Amount Spent'],outputCol='features')
output=assemble.transform(data)
output.head(1)
```

```python
Final_data=output.select(['features','Yearly Amount Spent'])
Final_data.show()
```

```python
train_data,test_data=Final_data.randomSplit([0.7,0.3])
train_data.describe().show()
```

```python
test_data.describe().show()
```

```python
lr=LinearRegression(labelCol='Yearly Amount Spent')
```

```python
lr_model=lr.fit(train_data)
```

```python
test_results=lr_model.evaluate(test_data)
```

```python
test_results.residuals.show()
```

```python
test_results.rootMeanSquaredError
test_results.r2
```
